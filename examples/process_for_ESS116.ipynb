```python
import xarray as xr
import numpy as np
```

```python
ds = xr.concat([
    xr.open_dataset(f"../data/coarsened/CM4Xp125_budgets_sigma2_{start}-{start+4}.zarr")[["siconc", "tos"]]
    for start in np.arange(2000, 2050, 5)
], dim="time")

for v in ds.data_vars:
    ds[v] = ds[v].fillna(0.).where(ds.wet)

for v in ["NCO", "associated_files", "external_variables", "filename", "history", "grid_tile", "title"]:
    del ds.attrs[v]
```

```python
ds
```

```python
ds.sel(exp="control").drop_vars(["exp"]).to_netcdf("../data/CM4X_piControl_1850.nc")
ds.sel(exp="forced").drop_vars(["exp"]).to_netcdf("../data/CM4X_SSP585.nc")
```
